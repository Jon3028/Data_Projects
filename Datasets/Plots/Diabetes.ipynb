# Graficas con plotly Scatter
#### Importar librerias 
#### **import plotly.offline as pyo**
#### **import plotly.graph_obj as go**

```python
import plotly.offline as pyo
import plotly.graph_objs as go
```

## Tendremos 4 pasos para generar un grafico con plotly
### 1.- vamos a generar un objeto que tendra los datos
<code>data1=[go.Scatter(x=columna_dataframe_x,y=columna_dataframe_y,resto de atributos)]</code>
### 2- vamos a generar un objeto con el layout de nuestro grafico
<code>layout1=go.Layout(title="titulo del grafico",xaxis=dict(title="Titulo eje X)",yaxis=dict(title="Titulo eje y"))</code>
### 3.- Mostraremos nuestro grafico con metodo figura
<code>fig=go.Figure(data=data,layout=layout)</code>
### 4.- Por ultimo llamaremos al metodo pyo el objeto fig y el nombre del fichero html del archivo
<code>pyo.plot(fig,filename="nombre_fichero.html")

```python
data1=[go.Scatter(x=data["Age"],y=data["BloodPressure"],mode="markers",marker=dict(size=12,symbol="circle",line={"width":3}))]
```

```python
layout1=go.Layout(title="Presion de sangre por edad",xaxis=dict(title="edad"),yaxis=dict(title="presion sanguinea"))
```

```python
fig=go.Figure(data=data1,layout=layout1)
```

```python
fig.show()
#Tenemos la opcion de generar un archivo HTML con la siguiente instruccion
#pyo.plot(fig,filename="presion_de_sangre.html")
```

# Creacion de graficos de barras
#### En el eje X vamos a colocar las categorias
#### En el eje Y vamos a colocar los valores

```python
import plotly.offline as pyo
import plotly.graph_objs as go


data1=[go.Bar(x=data["Age"],y=data["BloodPressure"])]
layout1=go.Layout(title="Presion de sangre por edad",xaxis=dict(title="Edad"),yaxis=dict(title="Presion de sangre"))
fig=go.Figure(data=data1,layout=layout1)
fig.show()
pyo.plot(fig,filename="Plots/Presion_por_edad.html")
```

### vamos a generar categorias de color para la presion mas alta


```python
import plotly.offline as pyo
import plotly.graph_objs as go

trace0=go.Bar(x=data["Age"],y=data[data["BloodPressure"]==100],name="Presion 100",marker={"color":"#FFD706"})
trace1=go.Bar(x=data["Age"],y=data[data["BloodPressure"]==90],name="Presion 90",marker={"color":"#82E0AA"})
trace2=go.Bar(x=data["Age"],y=data[data["BloodPressure"]==60],name="Presion 60",marker={"color":"#8E44AD"})

data1=[trace0,trace1,trace2]
#data1=[go.Bar(x=data["Age"],y=data["BloodPressure"])]
layout1=go.Layout(title="Presion de sangre por edad",xaxis=dict(title="Edad"),yaxis=dict(title="Presion de sangre"),barmode="stack")
fig=go.Figure(data=data1,layout=layout1)
fig.show()
pyo.plot(fig,filename="Plots/Presion_por_edad.html")
```
